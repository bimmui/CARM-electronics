```python
# importing necessary packages
# doing this first line bc using juypter notebook in vscode
import numpy as np
%matplotlib ipympl
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.widgets import CheckButtons
import pandas as pd
import scipy.stats as s_stats
import seaborn as sns
import mplcursors
from mpl_toolkits.mplot3d import Axes3D
```

```python
sns.set_theme(context='notebook', style='darkgrid', palette='colorblind', 
              font='sans-serif', font_scale=1, color_codes=True, rc=None) 
palette = sns.color_palette("colorblind")
```

```python
uncalibrated_data = pd.read_csv('UNCAL.CSV')

x_accel = uncalibrated_data['Accel_X']
y_accel = uncalibrated_data['Accel_Y']
z_accel = uncalibrated_data['Accel_Z']

x_mag = uncalibrated_data['Mag_X']
y_mag = uncalibrated_data['Mag_Y']
z_mag = uncalibrated_data['Mag_Z']

x_gyro = uncalibrated_data['Gyro_X']
y_gyro = uncalibrated_data['Gyro_Y']
z_gyro = uncalibrated_data['Gyro_Z']
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_accel, y_accel, z_accel, c='r', marker='o')


ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Raw Accelerometer Data in 3D Space')

# Show plot
plt.show()
```

```python
plt.figure()

plt.scatter(x_accel, y_accel, s=1, label='X/Y plane')
plt.scatter(x_accel, z_accel, s=1, label='X/Z plane')
plt.scatter(y_accel, z_accel, s=1, label='Y/Z plane')

plt.title('Raw Accelerometer Data')

plt.legend()

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_mag, y_mag, z_mag, c='r', marker='o')


ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Raw Magnetometer Data in 3D Space')

# Show plot
plt.show()
```

```python
plt.figure()

plt.scatter(x_mag, y_mag, s=1, label='X/Y plane')
plt.scatter(x_mag, z_mag, s=1, label='X/Z plane')
plt.scatter(y_mag, z_mag, s=1, label='Y/Z plane')

plt.title('Raw Magnetometer Data')

plt.legend()

plt.show()
```
